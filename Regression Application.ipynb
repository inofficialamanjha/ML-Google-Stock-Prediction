```python
import pandas as pd # pandas for managing csv file dataset
import quandl, math, datetime # quandl for dataset
import numpy as np # computing library for arrays, matrix
import matplotlib.pyplot as plt # plt is a matplotlib plot
from matplotlib import style
import pickle # pickle is serialization of any python object

style.use('ggplot') # set variable of the style_matplotlib object

# used for scalling, to increase accuracy, processing speed
# cross-validation to split in to training and testing samples, shuffles our data for us

from sklearn import preprocessing, svm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
## df = quandl.get('WIKI/GOOGL') : for getting data_set directly from quandl

df = pd.read_csv('WIKI_GOOGL.csv',header=0,index_col = 'Date',parse_dates = True)
```

Find correlation and relationships between columns.

In the dataset, acquire as much meaningful features you can get.

```python
# columns = ["Date","Open","High","Low","Close","Volume","Ex-Dividend","Split Ratio","Adj. Open","Adj. High","Adj. Low","Adj. Close","Adj. Volume"]

df
```

```python
# panda's is reading the dataframe in ascending order of data

# to convert data_frame in descending order of date

df = df.sort_index(ascending=False)

df = df.reset_index() # resents the index
```

```python
df
```

```python
# recreate data-frame to have minimum features

df = df[['Adj. Open','Adj. High','Adj. Low','Adj. Close','Adj. Volume']]

df
```

```python
# we are making new features and removing the redundant once

df['HL_PCT'] = (df['Adj. High'] - df['Adj. Low']) / df['Adj. Close'] * 100 # create a new column

df['PCT_change'] = (df['Adj. Close'] - df['Adj. Open']) / df['Adj. Open'] * 100

df = df[['Adj. Close','HL_PCT','PCT_change','Adj. Volume']]

# columns = ["Date","Adj. Close","HL_PCT","PCT_change","Adj. Volume"]

df
```

```python
forecast_col = 'Adj. Close' # forecast column

df.fillna(-99999, inplace=True) # fillna stands for not an available nan. Since in ML we can't work with nan data

forecast_out = int(math.ceil(0.01*len(df))) # we are going to predict 1% previous forecast

df['label'] = df[forecast_col].shift(forecast_out) # we are shifting the values forecast out values down by 1% the length of our data_frame

# the data frame with a new_column label

df 
```

```python
# define X features; labels Y

X = np.array(df.drop(['label'],1)) # df.drop returns a new data_frame

# scaling X feature matrix

X = preprocessing.scale(X) # scale along side other values

# taking X where 'lable' ( forecast_out) is NaN **&** taking X_lately to predict them later on

X = X[forecast_out:]

X_lately = X[:forecast_out]
```

```python
df.dropna(inplace=True) # drops the whole row, wherever one value in the row-column is nan

Y = np.array(df['label'])

print(len(X),len(Y))

print("\n")

print(df.tail(),"\n\n",df.head()) # viewing the data_frame from head to tail
```

```python
# can use cross_validation.text_train_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2) # divide set into 80:20 ration for training:testing; returns a tuple which is needed 
```

```python
# create a machine Learning model based on linear regression

model = LinearRegression(n_jobs = -1) # n_jobs represents threading

# fit the features and lables
model.fit(X_train,Y_train)
```

```python
# Pickling, scaling

# Pickling : Saving the model

with open('LinearRegression.pickle','wb') as f:
    pickle.dump(model,f) # to dump or to save the trained classifier

#load the classifier

pickle_in = open('LinearRegression.pickle','rb')
loaded_model = pickle.load(pickle_in) # load the saved model
```

```python
accuracy = loaded_model.score(X_test,Y_test) # calculate's accuracy

print(accuracy) # got 97.9% accuracy
```

```python
# Using Support vector Regression

svm_model = svm.SVR(kernel = 'poly') # default is a linear kernel

svm_model.fit(X_train,Y_train)

svm_accuracy = svm_model.score(X_test,Y_test)

print(svm_accuracy) # svm has got accuracy of 80% with linear kernel and 65.2% with polynimial kernel
```

```python
# prediction based on the X_lately set

forecast_set = model.predict(X_lately)

print(forecast_set, accuracy, forecast_out)
```
